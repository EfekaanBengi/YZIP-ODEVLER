```python
df3 = pd.read_csv("HousingData.csv")
```

```python
df3.isnull().sum()
```

```python
for col in df3.columns:
    if df3[col].dtype == "object":
        df3[col] = pd.to_numeric(df3[col])
```

```python
df3.interpolate(method="linear")
```

```python
df3.isnull().sum()
```

```python
df3.fillna(df3.median(numeric_only=True))
```

```python
df3.isnull().sum()
```

```python
df3.to_csv("Cleaned_HousingData.csv")
```
